#### **dict**<br/>
Es una colección de par clave-valor de tamaño flexible, donde la clave y el valor son objetos de Python. Para crear un objeto tipo dict se utilizan llaves {} y dos puntos para separar la clave y el valor:

````python
empty_dict = {}
print(empty_dict)
dict1 = {'a' : 'some value', 'b' : [1, 2, 3, 4]}
print(dict1)
````

Se puede acceder, insertar o adicionar elementos utilizando la misma sintaxis que para acceder a los elementos de una lista:

````python
## Adicionar clave y valor en un dict -> la clave 7 y valor 'an integer'
dict1[7] = 'an integer'
dict1['dummy'] = 'another value'
print(dict1)
## Imprimir el valor de una clave
print(dict1['a'])
##Verificar si un dict contiene una clave
print('b' in dict1)
````

````python
### combinar dos dict
dict2={'b' : 'foo', 'c' : 12}
dict1.update(dict2)
print(dict1)
````

````python
### imprimir las claves de un dict
print(dict1.keys())
### imprimir los valores de un dict
print(dict1.values())
````

````python
### Eliminar un elemento del dict con la palabra 'del' e indicando la clave
del dict1['b']
print(dict1)
### Eliminar un elemento del dict con la palabra 'pop' indicando la clave, esta instrucción retorna el valor de la clave borrada
elem_del = dict1.pop('dummy')
print(elem_del)
````

### **Functional programming:**
"Functional programming is a paradigm where the program flow is achieved mainly
through the evaluation of (mathematical) functions rather than through a series of steps that change the state of the program. Purely functional programs avoid the changing of state (side effects) and the use of mutable data structures."<br/>
"Functional programming is a programming paradigm in which the primary method of computation is evaluation of functions."

**Funciones Lambda**<br/>
Son un concepto de programación muy popular que es especialmente usado en programación funcional. En otros lenguajes de programación, Las funciones lambda se conocen como funciones anónimas, expresiones lambda o
literales de función. Las funciones Lambda son funciones anónimas que no tienen que ser vinculadas a un identificador (variable).<br/>
Las funciones lambda en Python solo se pueden definir mediante expresiones. La sintaxis de las funciones lambda son las siguientes:
```
# lambda <arguments>: <expression>
```




````python
## Función definida de forma normal
import math
def circle_area(radius):
  return math.pi * radius ** 2
````

````python
circle_area(42)
````

````python
circle_area
````

````python
## Equivalente en función lambda
circle_area=lambda radius: math.pi * radius ** 2
````

````python
circle_area(42)
````

````python
circle_area
````

````python
cube = lambda x: x * x*x 
print(cube(7))
````

````python
(lambda x1, x2, x3: (x1 + x2 + x3) / 3)(9, 6, 6)
````

````python
numeros = [2, 5, 10, 23, 50, 33]
list( map(lambda x: x*2, numeros) )
````

````python
S1 = ['a','b','c']
S2 = ['x','y','z']
list(map(lambda x, y:x+y,S1,S2))
````
